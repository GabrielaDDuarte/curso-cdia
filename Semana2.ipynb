Agregação

```python
#groupby vai agregar qualitativa e quantitativa
df = pd.read_csv('Banco de Dados Laboratorio.csv')
```

```python
df.groupby(['STATUSCIVIL'])['IDADE'].mean()
```

```python
#Quero agrupar por status civil e genero:
df.groupby(['STATUSCIVIL','GENERO'])['IDADE'].mean()
```

```python
#Quero agregar múltiplas métricas:
df.groupby(['STATUSCIVIL']).agg({
                                'RENDAMENSAL':['sum','mean','count'],
                                'IDADE':['mean']
})
```

```python
#Quero ver a diferença por cargo:
df.groupby(['CARGO']).agg({
                        'RENDAMENSAL':['sum','mean','count'],
                        'IDADE':['mean']
})
```

```python
#A tabela dinamica é um pivot. Eu posso pegar meu filtro (cargo, por exemplo) e transformar em novas colunas:
df.pivot_table(values='IDADE',
              index='CARGO',
              columns='GENERO',
              aggfunc='mean')
```

```python
#Para fazer a contagem dos dados

df.pivot_table(values='IDADE',
              index='CARGO',
              columns='GENERO',
              aggfunc='count')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cargo_cont = df['CARGO'].value_counts(dropna=False)
```

```python
cargo_cont
```

```python
plt.bar(cargo_cont.index,cargo_cont.values)
```

```python
plt.barh(cargo_cont.index,cargo_cont.values)
plt.title('Quantidade Cargo')
plt.ylabel('Quantidade')
plt.xlabel('Cargo')
plt.xticks(rotation=80)

plt.show()
```

Análise Quantitativa

```python
#Intervalo interquartilico
Q1 = df['ANONAEMPRESA'].quantile(0.25)
Q3 = df['ANONAEMPRESA'].quantile (0.75)

#Calcular o IQR
IQR = Q3 - Q1

#Calcular limites para outlier
limite_inf = Q1 - 1,5 * IQR
limite_sup = Q3 + 1,5 * IQR
```
